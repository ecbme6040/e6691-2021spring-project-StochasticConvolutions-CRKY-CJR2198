# Data Augmentation Illustration Notebook


In this notebook I illustrate the augmentation for the images in both high and low resolution using the custom dataloader.

```python
import pandas as pd
import numpy as np
from torch.utils.data import Dataset 
from torch.utils.data import DataLoader
from model_files.utils import *
from matplotlib import pyplot as plt
```

```python
#Create Dataset
train_df = pd.read_csv('data/1024x1024_train.csv')
train_dataset = MelaDS(image_folder='data/1024x1024/', dataframe=train_df, global_res=224,high_res=1024)
train_dataset = DataLoader(train_dataset)
iterator = iter(train_dataset)
```

```python
#find a malignant image
%matplotlib inline
target = 0.0
while target == 0.0:
    batch = next(iterator)
    lr_img, hr_img, target = batch
```

```python
## Display low resolution augmented image
npimg = lr_img[0].numpy()
plt.figure(figsize=(2.5,2.5))
plt.imshow(np.transpose(npimg, (1,2,0)), interpolation='nearest')
```

```python
## Display high resolution augmented image
npimg = hr_img[0].numpy()
plt.figure(figsize=(6,6))
plt.imshow(np.transpose(npimg, (1,2,0)), interpolation='nearest')
```
